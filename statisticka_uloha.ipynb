```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv('children_per_woman_total_fertility.csv')
data
```

```python
columns=['country']
columns += [str(i) for i in range(1800, 2022)]
actual_data = data[columns].copy()
actual_data.sort_values(by='country', inplace=True)
actual_data
```

## Tranforming data structure a bit so it's more convenient to work with

```python
formated_data = actual_data.transpose().reset_index()
formated_data.columns = formated_data.iloc[0]
formated_data = formated_data[1:].copy().rename(columns={'country': 'year'})
formated_data['year'] = formated_data['year'].apply(lambda x: int(x))
formated_data
```

## Let's take a look at the number of children per year per each country

```python
countries = formated_data.columns[1:]
fig = px.scatter(formated_data, x='year', y=countries)
fig.show()
```

## The impact of the WW2 on the fertility of countries which took big part in WW2 [linear regression on years 1935 - 1946]

```python
from sklearn.linear_model import LinearRegression
# No data for England and Slovakia
ww2_countries=['Poland', 'France', 'Italy', 'Japan', 'Finland', 'Germany',
               'Austria', 'Ukraine', 'Russia', 'Czech Republic', 'Belgium', 'Netherlands', 'Luxembourg', 'USA']
ww2_countries.sort()
ww2_data = formated_data[(formated_data['year'] >= 1935) & (formated_data['year'] <= 1946)].copy()
x = ww2_data[['year']]
for country in ww2_countries:
    cur_data = ww2_data[['year', country]].copy()
    cur_data[country] = cur_data[country].apply(lambda x: float(x))
    y_country = cur_data[country]
    model = LinearRegression()
    model.fit(x, y_country)
    y_pred = model.predict(x)
    cur_data['linear_'+country] = y_pred
    fig = px.scatter(cur_data, x='year', y=[country])
    fig.add_scatter(x=cur_data['year'], y=cur_data['linear_'+country], mode='lines', name='trendline')
    fig.show()
```

### Austria, Poland, Ukraine, Russia show the decrease of children per year during WW2 and for some time after it. Linear regression for Austria doesn't reflect it because of the big increase in 1939 compared to previous years.

## Check whether the average fertility per country in the year 2021 is a normal distribution throughout the whole world with Shapiro-Wilk test. The null hypothesis is that the data is from the normal distribution.

```python
import numpy as np
from scipy.stats import shapiro

data2021 = formated_data[formated_data['year'] == 2021].dropna(axis=1)
children2021 = data2021.iloc[0][1:].tolist()
children2021.sort()
children2021
```

```python
shapiro_test = shapiro(children2021)
shapiro_test
```

### p-value is 1e-12, so the null hypothesis is rejected, which means that the data isn't from the normal distribution.

## Check the same thing, but only for European countries

```python
# Slovakia and Cyprus aren't in the dataframe, so I don't include them here.
eu_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Czech Republic',
    'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland',
    'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Poland',
    'Portugal', 'Romania', 'Slovenia', 'Spain', 'Sweden'
]
data2021_eu = data2021[eu_countries]
data2021_eu
```

```python
children2021_eu = data2021_eu.iloc[0].tolist()
children2021_eu.sort()
children2021_eu
```

```python
shapiro_test_eu = shapiro(children2021_eu)
shapiro_test_eu
```

### p-value is much more than 0.05, so the data can be from a normal distribution

### let's take a look at the average value of fertility from averages per country in Europe and the standard deviation:

```python
mean = np.mean(children2021_eu)
sd = np.std(children2021_eu)
mean, sd
```
